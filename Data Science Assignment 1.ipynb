```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('sample_data_50_rows.csv')
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.info()
```

**Filling Missing Data**

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data['Salary'] = data['Salary'].fillna(data['Salary'].mode()[0])
data['Department'] = data['Department'].fillna('Accounting')
data['Sales'] = data['Sales'].fillna(data['Sales'].mean())

data.head(10)
```

**Encoding the categorical data (Department, Education)**

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data['Department'] = lb.fit_transform(data['Department'])
data['Education'] = lb.fit_transform(data['Education'])
data.head(10)
```

**Visualization using Bar Chart, Histogram, Box Plot**

```python
plt.figure(figsize=(7,6))
sns.countplot(x = 'Age', data = data)
plt.title('Bar Chart showing the frequency of Age' )
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(7,6))
data['Salary'].plot(kind = 'hist', bins = 8, edgecolor = 'red')
plt.title('Bar Chart showing the frequency of Salary relatively to the Employee')
plt.xlabel('Salary')
plt.ylabel('No of Employee')
plt.show()
```
